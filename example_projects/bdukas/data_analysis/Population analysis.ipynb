```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

I began by loading in and making basic observations about the population dataset.

```python
pop_df = pd.read_csv('../data/Raw/world_pop_1000s.csv')
pop_df.shape
```

```python
pop_df.sample(12)
```

I then dropped two irrelevant columns.

```python
pop_df2 = pop_df.drop(columns=['Note', 'Sex'])
pop_df2.shape
```

```python
pop_df2.head()
```

I dropped rows of the data that lacked data. This seemed to be only four rows.

```python
pop_clean_df = pop_df2.dropna()
```

```python
pop_clean_df.shape
```

```python
pop_clean_df.head()
```

```python
pop_clean_df.columns
```

I created a label for all of the year columns, which I used to calculate the average population for a country over the years represented in the data.

```python
year_cols = [str(y) for y in range(1990,2019)]
```

```python
pop_clean_df['Average']=pop_clean_df[year_cols].sum(axis=1) / len(year_cols)
```

```python
pop_clean_df['Average']
```

```python
pop_clean_df.head()
```

I renamed several countries to match my other data.

```python
pop_clean_df['Location'] = pop_clean_df['Location'].replace({'Bolivia (Plurinational State of)':'Bolivia', 
                      'Brunei Darussalam':'Brunei', 'Cabo Verde':'Cape Verde', 'Côte d’Ivoire':"Cote d'Ivoire",
                            'Czechia':'Czech Republic', "Democratic People's Republic of Korea":'North Korea'})
```

```python
pop_clean_df['Location']=pop_clean_df['Location'].replace({'Eswatini':'Swaziland', 'Iran (Islamic Republic of)':'Iran',
                                      "Lao People's Democratic Republic":'Laos', 'Micronesia (Fed. States of)':'Micronesia (country)',
                                      'Republic of Korea':'South Korea', 'Republic of Moldova':'Moldova',
                                       'Russian Federation':'Russia'})
```

```python
pop_clean_df['Location']=pop_clean_df['Location'].replace({'State of Palestine':'Palestine', 'Syrian Arab Republic':'Syria',
                                      'Timor-Leste':'Timor',
                                      'United Republic of Tanzania':'Tanzania', 'United States of America':'United States',
                                      'Venezuela (Bolivarian Republic of)':'Venezuela', 'Democratic Republic of the Congo':'Congo'})
```

Finally, I exported the cleaned data to a new CSV.

```python
pop_clean_df.to_csv('../data/Cleaned/world_pop_1000s_clean', index=False)
```
