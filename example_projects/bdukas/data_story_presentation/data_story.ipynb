## Introduction

Mental illness is a major global health issue. Around the world, individuals face a variety of mental health problems. Important in the treatment of these issues are mental health professionals. Specifically, psychiatrists and psychologists are vital resources for those battling mental illness. Unfortunately, many choose to forgo treatment and suffer in silence due to the stigma surrounding mental illness.

Using data from the World Health Organization and the Institute for Health Metrics and Evaluation, this data story aims to compare rates of mental disorders around the world and observe how these rates have changed over time. It will compare these rates by continent, and consider the availability of mental health professionals in relation to the rates of disorders. This will give insight into how many people are affected by mental illness, where they are, and which of them have access to the most resources.

## Definitions

Before discussing the data, it is important to define a few key terms.

**Mental and Substance Use Disorders**: This term includes a variety of complex mental health problems that can be diagnostically characterized as a disorder. Examples include anxiety, depression, bipolar disorder, schizophrenia, and eating disorders. Also included are substance use disorders, such as alcohol and drug addiction.

**Anxiety**: For the purposes of this data story, anxiety is used in reference to anyone with a clinically diagnosable anxiety disorder. This includes a range of disorders, including generalized anxiety disorder, post-traumatic stress disorder (PTSD), obsessive-compulsive disorder (OCD), and social anxiety disorder, among others. These are categorized by a range of symptoms, and defined by worry/apprehension and autonomic overreactivity.

**Depression**: For the purposes of this data story, depression is used in reference to anyone with a clinically diagnosable depressive disorder, such as severe depressive disorder and mild persistent depressive disorder. These are categorized by a range of symptoms, but defined by diminished mood and negative affect.

**Psychiatrist**: A physician who specializes in the prevention, diagnosis, and treatment of mental illness (Shiel, 2018). These individuals can prescribe medication for mental illness and treat through a variety of psychotherapies. 

**Psychologist**: A professional specializing in diagnosing and treating diseases of the brain, emotional disturbance, and behavior problems (Shiel, 2018). Psychologists treat individuals through a variety of psychotherapies, but cannot prescribe medication.


**Human resources**: This a term used to reference psychiatrists and psychologists, who are human resources for individuals with mental disorders.

**Stigma**: a mark of disgrace associated with a particular circumstance, quality, or person. This term is often used in discussing mental illness.

For the purposes of this data story, "more developed, Westernized, and wealthier countries" are (usually) those in North America, Europe, and Oceania, while "less developed, less Westernized, poorer" countries are those in Asia and Africa, with South American countries falling somewhere in between. This categorization is substantiated by the data in Figure 1 from the International Monetary Fund.

## Data background

This data story relies on four major data sets. 

Three of these come from the Institute for Health Metrics and Evaluation (IHME)'s Global Burden of Disease study. They measure (1) rates of mental and substance use disorders around the world, (2) rates of anxiety  around the world, and (3) rates of depression around the world. The data is "collected and analyzed by a consortium of more than 3,600 researchers in more than 145 countries" (IHME, 2018). It includes information for every year from 1990 to 2017 for 231 unique geographic entities.

A fourth dataset comes from the World Health Organization. It measures the number of human mental health resources in 154 countries. Each country's data is from a different year between 2013-2016. 
A fifth dataset about suicide was included in the initial analysis but excluded from this data story in order to maintain a reasonable length.

## Hypotheses

Several hypotheses drive this analysis. Hypothesis 1 regards the rates of mental disorders around the world. Hypothesis 2 discusses change of mental illness rates over time. Hypothesis 3 is focused on the ratio of treatment to disorder prevalence.

**H1.0**: The rates of mental disorders will be roughly the same around the world.

**H1.1**: More developed, westernized countries will display higher rates of mental disorders.

**H1.2**: Less developed, less westernized countries will display higher rates of mental disorders

**H2.0**: The rates of mental disorders will remain the same over time. 

**H2.1**: The rates of mental disorders will decrease over time.

**H2.2**: The rates of mental disorders will increase over time.

**H3.0**: Disordered individuals have equal access to psychiatrists and psychologists, regardless of country.

**H3.1**: Disordered individuals from more developed, Westernized countries will have more access to psychiatrists and psychologists than those in less developed, poorer countries.

**Note**: This series of hypotheses  can be proven or disproven observationally. The author was careful not to attribute a cause to any of these. In the various discussion sections, he conjectures about potential reasons for the observations while recognizing that these remain unsubstantiated by this research.

## Method

Each of these datasets was downloaded and cleaned. The data was merged with other data that contained each country's population and continent. This required relabelling of many countries to successfully merge the various datasets (for example, United States of America was switched to United States, Eswatini to Swaziland, etc.). Information from many territories that are not official countries were excluded from analysis (though some remain) based on technicalities of the data set. Using the rates of disorders (in percent) and the total population of the country, the author mathematically calculated the number of disordered individuals in each country. Eventually, the three datasets from IHME were merged together for analysis. 186 countries were included in this merged dataset. The data was used to generate maps of disorder prevalence, compare rates by continent, and observe changes over time.

The WHO dataset about human resources was then merged with the previous three. The result of this merge brought the number of countries down to 120. Guyana was excluded from this analysis since the data indicated it had zero psychologists, which is not in fact true. This data was then used to assess the ratios of human resources to populations with mental disorders.

## General Discussion

The data analysis shows that more westernized and developed countries tend to display higher rates of mental disorders than less developed, less westernized countries, with the unique exception of the Arab/Muslim world. Rates of disorders overall have remained consistent over the past three decades, though rates in Africa and Asia have increased more in comparison to the other continents.

The nature of this data does not lend itself to causal explanations. Thus, interpretations of these trends are based on conjecture and the research of others.

It might be tempting to find geopolitical explanations for the varying rates of disorders by location. However, this data likely does not accurately reflect the actual prevalence of mental disorders; instead, it shows the *reports* of mental disorders around the world. **Stigma** is a major buzz word used when describing the conversation surrounding mental illness. For a variety of reasons, many individuals choose to keep their mental illnesses secret, often due to embarrassment and shame surrounding mental illness. 

In their study "World survey of mental illness stigma," Seeman et. al (2016) found stigma surrounding mental illness is higher in developing countries. Those in developing countries are more likely than those in developed countries to believe that individuals with mental illnesses are violent. While those in developed countries often view mental illness as similar to a physical disease, those in developing countries tend to view it as "supernatural and mystical, punishment for a person’s sins or those of ancestors."

These findings may explain the data in this data story. Individuals in more developed, Westernized countries feel less pressure from stigma and are more likely to report mental illness, while those in less developed, less Westernized countries submit to the stigma and do not report their illnesses. The data in this data set is a measurement of rates of mental illness around the world, but it may really reflect stigma in individual countries of the world. The higher increase of the rates in Africa and Asia may be the result of destigmatization over time in these areas and thus higher reports of mental illness. This could explain the high rates of change observed in countries of the Arab world (such as in Qatar and UAE), who have become increasingly developed and westernized. Another explanation for this trend might increasingly better measures of data collection in less developed countries.

More developed, Westernized countries have many more psychiatrists and psychologists than less developed, less Westernized ones. This might suggest that those in more developed countries have better access to treatment and are more likely to recover from their mental illnesses. However, the prevalence data does not indicate that this is the case, as westernized, developed countries still tend to show high rates of mental illness. We are therefore unable to assess the effectiveness of psychiatrists and psychologists from this data. Theoretically, it is plausible to say that disordered individuals in more developed countries have more access to treatment resources than those in less developed countries. However, the ratio of psychiatrists or psychologists to disordered individuals does not accurately reflect access. There are many barriers to treatment based on cost, insurance policies, and geographic location that may prevent individuals from seeking treatment, even if their country has many mental health professionals. For example, the disordered population in a country might be distributed throughout rural and urban areas, while the psychiatrists and psychologists are concentrated in the urban areas. Accessing treatment then becomes very difficult for certain members of the the population.

## Limitations

Several limitations affect this data analysis. As already mentioned, issues of stigma bias the rates of report of mental disorders in different countries, and causality cannot be established. Similarly, the World Health Organization and the Institute for Health Metrics and Evaluation were vague about their data collection methods, and it is difficult to assess how accurate it all is. It may be that mental illness was underreported in some countries simply because WHO and IHME lacked the resources get an accurate assessment, especially in less developed countries. Additionally, a lot of data was missing, especially in the psychiatrists/psychologists data; only 115 countries' numbers of psychiatrists were available, and only 90 countries' numbers of psychologists were available. There are 193 member states of the United Nations, so much data was missing. Alternatively, the author is a novice programmer, and thus some countries were lost in the various merges and transfers of the data.

It is not clear from the data exactly what the threshold was for determining whether an individual has a mental disorder. While the American Psychological Association and the World Health Organization have official definitions and diagnostic criteria, these might not be used consistently around the world. It may be that the standards of diagnosis are higher or lower in different countries, thus skewing the data. A last limitation is that this data story relies heavily on visualization as opposed to statistics. This makes it easily digestible, but challenging to develop more nuanced conclusions.

## Conclusion & Directions for Further Study

Mental disorders, and specifically anxiety and depressive disorders, still riddle the planet and are fairly consistently geographically distributed, with a bias towards more developed, Westernized countries and Arab countries. There has been little change over the past three decades, though an increase can be observed in African and Asian countries. The reason for these trends is hard to discern, but it is clear that mental illness is a global issue and requires more research, planning, and solutions to mitigate. Psychiatrists and psychologists are much more numerous in developed countries than less developed ones, with African countries having the most major shortages of treatment providers.

Future studies could attempt to determine causal factors for the findings of this data story. It is important to understand why reported rates of mental disorders have increased in some regions while staying stagnant or decreasing in other areas. It could also look at factors in individual countries that affect people's likelihood to seek treatment, assessing whether resource-to-disorder ratios are accurate measures of treatment availability. These studies should have more apparent methods of data collection and defined diagnostic criteria.
