```python
#which is the best wharton concentration?
#let's look at average course quality
#first we'll need to break up the 'section' column because we can't compare ACCT to STAT etc
#we'll make one column for department and one for course number
#and we'll have to define two functions to do so:
```

```python
def find_section_no (section):
    return(''.join(ch for ch in section if not ch.isalpha()))
```

```python
def find_dept (section):
    return(''.join(ch for ch in section if ch.isalpha()))
```

```python
wharton19A['department'] = wharton19A['section'].apply(find_dept)
```

```python
wharton19A['section_no'] = wharton19A['section'].apply(find_section_no)
```

```python
new_wharton19A=wharton19A[['department','section_no','term','section','title','instructor','course_qual',
                     'instructor_qual','instructor_comm','stimulate',
                     'instructor_access','readings_qual','amount_learned',
                     'difficulty','workload','recommend_major',
                      'recommend_nonmajor']]
```
