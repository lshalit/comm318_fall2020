```python
import pandas as pd
```

* Import data to notebook

```python
favorite_platforms = pd.read_csv('../data/raw/favorite-social-networks-2012-2019.csv')
```

```python
favorite_platforms
```

* Change column names

```python
favorite_platforms.columns = ['Time Period', 'Snapchat', 'Instagram', 'Twitter', 'Facebook', 'Tumblr', 'Pinterest', 'Google+','Other', 'Percent']
```

```python
favorite_platforms
```

* Delete Percent column

```python
favorite_platforms.drop(["Percent"], axis = 1, inplace = True) 
```

```python
favorite_platforms
```

* Delete first two rows

```python
favorite_platforms.drop([0,1], axis=0, inplace=True)
```

```python
favorite_platforms
```

* Reset row numbers

```python
favorite_platforms.reset_index(drop=True, inplace=True)
```

```python
favorite_platforms
```

* Save to clean folder

```python
favorite_platforms.to_csv('favorite_platforms_clean.csv', index=False)
```
