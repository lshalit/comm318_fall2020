```python
import pandas as pd
```

* Import data

```python
data = pd.read_csv('../data/clean/favorite_platforms_clean2.csv')
```

```python
data
```

* Graph data

```python
import matplotlib.pyplot as plt
```

```python
data.set_index("Time Period",drop=True,inplace=True)
```

```python
data.plot(kind='bar', figsize=(20,10), fontsize=13, stacked=True)
plt.title('U.S. Teens Favorite Platforms 2012-2019', fontsize=16)
plt.xlabel('Time Period')
plt.ylabel('Percentage')
plt.tight_layout()
plt.savefig('Teens_Favorite_Platforms_2012_2019_Graph.png')
```
