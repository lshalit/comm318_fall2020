```python
import pandas as pd
```

* Import dataframe

```python
recruiters = pd.read_csv('../data/clean/social_media_recruiters_clean.csv')
```

```python
recruiters
```

```python
import matplotlib.pyplot as plt
```

* Graph data

```python
recruiters.set_index("Damaging Discoveries",drop=True,inplace=True)
```

```python
recruiters.plot(kind='barh', figsize=(10,5), fontsize=13, color='orange')
plt.title('Damaging Discoveries on Social Media Found by Recruiters', fontsize=16)
plt.xlabel('Percentage', fontsize=14)
plt.ylabel('Discovery', fontsize=14)
plt.tight_layout()
plt.savefig('Recruiters_Graph.png')
```
