# Currency

## Set Up

```python
%matplotlib inline

import warnings
warnings.simplefilter('ignore')

import pandas as pd
import matplotlib.pyplot as plt
```

## Load Data

```python
# Load in big data set
df = pd.read_csv('../data/whole_GEI_dataset.csv')
```

## Analyze

### 1. Gender Equality by National Currency

```python
# Create separate dataframes for countries that use the euro and ones that use local currency
euro_cos = df[df['Eurozone'] == 'Yes']
local_cos = df[df['Eurozone'] == 'No']

# Get the average GEI for each
euro_avg = euro_cos.groupby(['Year'])['Gender Equality Index'].mean()
local_avg = local_cos.groupby(['Year'])['Gender Equality Index'].mean()

# Convert to lists
euro = euro_avg.values.tolist()
local = local_avg.values.tolist()
```

```python
# Create lists of years and currency areas to use in plot
years = ['2005','2010','2012','2015']
currencies = ['Nations Using the Euro', 'Nations Using Local Currencies']
```

```python
# Plot lines for each on a graph
fig, ax = plt.subplots()
ax.plot(years, euro, color='mediumblue')
ax.plot(years, local, color='forestgreen')

# Add labels to the graph
plt.xlabel('Year',  size=12);
plt.ylabel('Gender Equality Index', size=12);
plt.title('Gender Equality by National Currency', size=14);
plt.legend(currencies, bbox_to_anchor=(1.05, .75))

# Save graph
plt.savefig('gei_currency.png')
```

### 2. Gender Equality in Money by National Currency

```python
# Get 'POWER' index averages for each currency group
m_euro_avg = euro_cos.groupby(['Year'])['MONEY'].mean()
m_local_avg = local_cos.groupby(['Year'])['MONEY'].mean()

# Convert to lists
m_euro = m_euro_avg.values.tolist()
m_local = m_local_avg.values.tolist()
```

```python
# Plot lines on a graph
fig, ax = plt.subplots()
ax.plot(years, m_euro, color='mediumblue')
ax.plot(years, m_local, color='forestgreen')

# Add labels to the graph
plt.xlabel('Year',  size=12);
plt.ylabel('Gender Equality "Money" Index', size=12);
plt.title('Gender Equality in Money by National Currency', size=14);
plt.legend(currencies, bbox_to_anchor=(1.65, .65))

# Save graph
plt.savefig('mon_currency.png')
```

### 3. Gender Equality in Financial Resources by National Currency

```python
# Get financial resources averages for each currency group
fr_euro_avg = euro_cos.groupby(['Year'])['Financial resources'].mean()
fr_local_avg = local_cos.groupby(['Year'])['Financial resources'].mean()

# Convert to lists
fr_euro = fr_euro_avg.values.tolist()
fr_local = fr_local_avg.values.tolist()

# Plot lines for each region on a graph
fig, ax = plt.subplots()
ax.plot(years, fr_euro, color='mediumblue')
ax.plot(years, fr_local, color='forestgreen')

# Add labels to the graph
plt.xlabel('Year',  size=12);
plt.ylabel('Gender Equality "Financial Resources" Index', size=10);
plt.title('Gender Equality in Financial Resources by National Currency', size=14);
plt.legend(currencies, bbox_to_anchor=(1.65, .7))
                                                                              
# Save graph
plt.savefig('finr_currency.png')
```

```python
# Get economic situation averages for each currency group
es_euro_avg = euro_cos.groupby(['Year'])['Economic situation'].mean()
es_local_avg = local_cos.groupby(['Year'])['Economic situation'].mean()

# Convert to lists
es_euro = es_euro_avg.values.tolist()
es_local = es_local_avg.values.tolist()

# Plot lines for each region on a graph
fig, ax = plt.subplots()
ax.plot(years, es_euro, color='mediumblue')
ax.plot(years, es_local, color='forestgreen')

# Add labels to the graph
plt.xlabel('Year',  size=12);
plt.ylabel('Gender Equality "Economic Situation" Index', size=10);
plt.title('Gender Equality in Economic Situation by National Currency', size=14);
plt.legend(currencies, bbox_to_anchor=(1.05, .7))

# Save graph
plt.savefig('ecos_currency.png')
```
