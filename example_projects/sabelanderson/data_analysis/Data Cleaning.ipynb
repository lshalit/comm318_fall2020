### Load Data

### Merge Data Frames

```python
# Sort data frames so the information lines up

con_info_df = con_info_df.sort_values('Country Short')

gei_2005_df = gei_2005_df.sort_values('Country')
gei_2010_df = gei_2010_df.sort_values('Country')
gei_2012_df = gei_2012_df.sort_values('Country')
gei_2015_df = gei_2015_df.sort_values('Country')
```

```python
con_info_df.head(5)
```

```python
# Reset the indexes 

con_info_df = con_info_df.reset_index(drop=True)

gei_2005_df = gei_2005_df.reset_index(drop=True)
gei_2010_df = gei_2010_df.reset_index(drop=True)
gei_2012_df = gei_2012_df.reset_index(drop=True)
gei_2015_df = gei_2015_df.reset_index(drop=True)
```

```python
gei_2005_df.head(5)
```

```python
con_info_df.head(5)
```

```python
# Attach country information data to year data frames

i_2005_df = pd.concat([gei_2005_df, con_info_df], axis=1)
i_2010_df = pd.concat([gei_2010_df, con_info_df], axis=1)
i_2012_df = pd.concat([gei_2012_df, con_info_df], axis=1)
i_2015_df = pd.concat([gei_2015_df, con_info_df], axis=1)
```

```python
i_2005_df.sample(5)
```

```python
# Merge year data frames

# Add year column to each data frame

i_2005_df['Year'] = '2005'

i_2010_df['Year'] = '2010'

i_2012_df['Year'] = '2012'

i_2015_df['Year'] = '2015'
```

```python
# Merge the 4 data frames

frames = [i_2005_df, i_2010_df, i_2012_df, i_2015_df]

gei_df = pd.concat(frames)
```

```python
# Save dataframe ("gei_df") to data folder as "whole_GEI_dataset"

gei_df.to_csv('../data/whole_GEI_dataset.csv', index=False)
```
