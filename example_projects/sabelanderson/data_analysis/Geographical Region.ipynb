# Geographical Region

### Set Up

```python
%matplotlib inline

import warnings
warnings.simplefilter('ignore')

import pandas as pd
import matplotlib.pyplot as plt
```

### Load Data

```python
# Load in big data set
df = pd.read_csv('../data/whole_GEI_dataset.csv')
```

```python
df
```

### Analyze GEI by Geographical Region

```python
# Create separate dataframes for each region

central_cos = df[df['Region'] == 'Central']
northern_cos = df[df['Region'] == 'Northern']
southeastern_cos = df[df['Region'] == 'Southeastern']
southern_cos = df[df['Region'] == 'Southern']
western_cos = df[df['Region'] == 'Western']
```

```python
# Get the average GEI for each region in each year

central_avg = central_cos.groupby(['Year'])['Gender Equality Index'].mean()
northern_avg = northern_cos.groupby(['Year'])['Gender Equality Index'].mean()
southeastern_avg = southeastern_cos.groupby(['Year'])['Gender Equality Index'].mean()
southern_avg = southern_cos.groupby(['Year'])['Gender Equality Index'].mean()
western_avg = western_cos.groupby(['Year'])['Gender Equality Index'].mean()


# Check lists

print(central_avg)
print(northern_avg)
print(southeastern_avg)
print(southern_avg)
print(western_avg)
```

```python
# Turn dataframes with averages into lists for each region

central = central_avg.values.tolist()
northern = northern_avg.values.tolist()
southeastern = southeastern_avg.values.tolist()
southern = southern_avg.values.tolist()
western = western_avg.values.tolist()


# Check lists

print(central)
print(northern)
print(southeastern)
print(southern)
print(western)
```

```python
# Create lists of years and regions to use in plot

years = ['2005','2010','2012','2015']
regions = ['Central', 'Northern', 'Southeastern', 'Southern', 'Western']
```

```python
# Plot lines for each region on a graph

fig, ax = plt.subplots()
ax.plot(years, central, color='lightslategrey')
ax.plot(years, northern, color='dodgerblue')
ax.plot(years, southeastern, color='orangered')
ax.plot(years, southern, color='m')
ax.plot(years, western, color='green')


# Add labels to the graph

plt.xlabel('Year',  size=12);
plt.ylabel('Gender Equality Index', size=12);
plt.title('Gender Equality Index Over Time by Geographic Region', size=14);
plt.legend(regions, bbox_to_anchor=(1.05, .75))


# Save graph

plt.savefig('gei_georeg.png')
```
