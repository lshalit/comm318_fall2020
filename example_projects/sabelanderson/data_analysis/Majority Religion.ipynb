# Majority Religion

## Set Up

```python
%matplotlib inline

import warnings
warnings.simplefilter('ignore')

import pandas as pd
import matplotlib.pyplot as plt
```

## Load Data

```python
# Load in data set
df = pd.read_csv('../data/whole_GEI_dataset.csv')
```

## Analyze

### 1. Gender Equality by National Majority Religion

```python
df['Religion'].unique()
```

```python
# Create separate dataframes for each region
orthodox_cos = df[df['Religion'] == 'Orthodox']
protestant_cos = df[df['Religion'] == 'Protestant']
catholic_cos = df[df['Religion'] == 'Catholic']
```

```python
# Get the GEI for each country in each religion in each year [removed]
# orthodox_cos_avg = orthodox_cos.groupby(['Year', 'Country'])['Gender Equality Index']
# protestant_cos_avg = protestant_cos.groupby(['Year', 'Country'])['Gender Equality Index']
# catholic_cos_avg = catholic_cos.groupby(['Year', 'Country'])['Gender Equality Index']

# Get the average GEI for each religion in each year
orthodox_avg = orthodox_cos.groupby(['Year'])['Gender Equality Index'].mean()
protestant_avg = protestant_cos.groupby(['Year'])['Gender Equality Index'].mean()
catholic_avg = catholic_cos.groupby(['Year'])['Gender Equality Index'].mean()
```

```python
# Turn dataframes with averages into lists for each region
orthodox = orthodox_avg.values.tolist()
protestant = protestant_avg.values.tolist()
catholic = catholic_avg.values.tolist()
```

```python
# Create lists to label plot
religions = ['Majority Eastern Orthox Nations', 'Majority Protestant Nations', 'Majority Catholic Nations']
years = ['2005','2010','2012','2015']

# Plot lines for each religion on a graph
fig, ax = plt.subplots()
ax.plot(years, orthodox, color='chocolate')
ax.plot(years, protestant, color='mediumturquoise')
ax.plot(years, catholic, color='darkmagenta')

# Add labels to the graph
plt.xlabel('Year',  size=12);
plt.ylabel('Gender Equality Index', size=12);
plt.title('Gender Equality by National Majority Religion', size=14);
plt.legend(religions, bbox_to_anchor=(1.05, .75))

# Save graph
plt.savefig('gei_religion.png')
```

### 2. Gender Equality in Power by National Majority Religion

```python
# Get 'POWER' index averages for each religion
p_orthodox_avg = orthodox_cos.groupby(['Year'])['POWER'].mean()
p_protestant_avg = protestant_cos.groupby(['Year'])['POWER'].mean()
p_catholic_avg = catholic_cos.groupby(['Year'])['POWER'].mean()

# Convert to lists
p_orthodox = p_orthodox_avg.values.tolist()
p_protestant = p_protestant_avg.values.tolist()
p_catholic = p_catholic_avg.values.tolist()

# Plot lines on a graph
fig, ax = plt.subplots()
ax.plot(years, p_orthodox, color='chocolate')
ax.plot(years, p_protestant, color='mediumturquoise')
ax.plot(years, p_catholic, color='darkmagenta')

# Add labels to the graph
plt.xlabel('Year',  size=12);
plt.ylabel('Gender Equality "POWER" Index', size=12);
plt.title('Gender Equality in Power by National Majority Religion', size=14);
plt.legend(religions, bbox_to_anchor=(1.65, .75))

# Save graph
plt.savefig('pow_religion.png')
```

### 3a. Gender Equality in Housework by National Majority Religion

```python
# Get housework numbers for each gender and religion
hw_orthodox_avg = orthodox_cos.groupby(['Year'])['People doing cooking and/or household, every day (%) W'].mean()
hw_protestant_avg = protestant_cos.groupby(['Year'])['People doing cooking and/or household, every day (%) W'].mean()
hw_catholic_avg = catholic_cos.groupby(['Year'])['People doing cooking and/or household, every day (%) W'].mean()
hm_orthodox_avg = orthodox_cos.groupby(['Year'])['People doing cooking and/or household, every day (%) M'].mean()
hm_protestant_avg = protestant_cos.groupby(['Year'])['People doing cooking and/or household, every day (%) M'].mean()
hm_catholic_avg = catholic_cos.groupby(['Year'])['People doing cooking and/or household, every day (%) M'].mean()

# Differences
hd_orthodox_avg = hw_orthodox_avg - hm_orthodox_avg
hd_protestant_avg = hw_protestant_avg - hm_protestant_avg
hd_catholic_avg = hw_catholic_avg - hm_catholic_avg

# Convert to lists
hw_orthodox = hw_orthodox_avg.values.tolist()
hw_protestant = hw_protestant_avg.values.tolist()
hw_catholic = hw_catholic_avg.values.tolist()
hm_orthodox = hm_orthodox_avg.values.tolist()
hm_protestant = hm_protestant_avg.values.tolist()
hm_catholic = hm_catholic_avg.values.tolist()

hd_orthodox = hd_orthodox_avg.values.tolist()
hd_protestant = hd_protestant_avg.values.tolist()
hd_catholic = hd_catholic_avg.values.tolist()

# Plot lines on a graph
fig, ax = plt.subplots()
ax.plot(years, hw_orthodox, color='chocolate')
ax.plot(years, hw_protestant, color='mediumturquoise')
ax.plot(years, hw_catholic, color='darkmagenta')
ax.plot(years, hm_orthodox, color='burlywood')
ax.plot(years, hm_protestant, color='paleturquoise')
ax.plot(years, hm_catholic, color='orchid')

# Add labels to the graph
gen_religions = ['Women in Majority Eastern Orthox Nations', 'Women in Majority Protestant Nations', 
                 'Women in Majority Catholic Nations', 'Men in Majority Eastern Orthox Nations', 
                 'Men in Majority Protestant Nations', 'Men in Majority Catholic Nations']

plt.xlabel('Year',  size=12);
plt.ylabel('% of People Doing Housework Daily', size=12);
plt.title('Gender Equality in Housework by National Majority Religion', size=14);
plt.legend(gen_religions, bbox_to_anchor=(1.05, .75))

# Save graph
plt.savefig('hosa_religion.png')
```

### 3b. Gender Equality in Housework by National Majority Religion

```python
# Differences graphs
fig, ax = plt.subplots()
ax.plot(years, hd_orthodox, color='chocolate')
ax.plot(years, hd_protestant, color='mediumturquoise')
ax.plot(years, hd_catholic, color='darkmagenta')

# Add labels to the graph
plt.xlabel('Year',  size=12);
plt.ylabel('% More Women than Men Doing Housework Daily', size=9);
plt.title('Gender Equality in Housework by National Majority Religion', size=14);
plt.legend(religions, bbox_to_anchor=(1.05, .75))

# Save graph
plt.savefig('hosb_religion.png')
```

### 4a. Gender Equality in Carework by National Majority Religion

```python
# Get carework numbers for each gender and religion
cw_orthodox_avg = orthodox_cos.groupby(['Year'])['People caring for and educating their children or grandchildren, elderly or people with disabilities, every day (%) W'].mean()
cw_protestant_avg = protestant_cos.groupby(['Year'])['People caring for and educating their children or grandchildren, elderly or people with disabilities, every day (%) W'].mean()
cw_catholic_avg = catholic_cos.groupby(['Year'])['People caring for and educating their children or grandchildren, elderly or people with disabilities, every day (%) W'].mean()
cm_orthodox_avg = orthodox_cos.groupby(['Year'])['People caring for and educating their children or grandchildren, elderly or people with disabilities, every day (%) M'].mean()
cm_protestant_avg = protestant_cos.groupby(['Year'])['People caring for and educating their children or grandchildren, elderly or people with disabilities, every day (%) M'].mean()
cm_catholic_avg = catholic_cos.groupby(['Year'])['People caring for and educating their children or grandchildren, elderly or people with disabilities, every day (%) M'].mean()

# Convert to lists
cw_orthodox = cw_orthodox_avg.values.tolist()
cw_protestant = cw_protestant_avg.values.tolist()
cw_catholic = cw_catholic_avg.values.tolist()
cm_orthodox = cm_orthodox_avg.values.tolist()
cm_protestant = cm_protestant_avg.values.tolist()
cm_catholic = cm_catholic_avg.values.tolist()

# Plot lines on a graph
fig, ax = plt.subplots()
ax.plot(years, cw_orthodox, color='chocolate')
ax.plot(years, cw_protestant, color='mediumturquoise')
ax.plot(years, cw_catholic, color='darkmagenta')
ax.plot(years, cm_orthodox, color='burlywood')
ax.plot(years, cm_protestant, color='paleturquoise')
ax.plot(years, cm_catholic, color='orchid')

# Add labels to the graph
plt.xlabel('Year',  size=12);
plt.ylabel('% of People Doing Carework Daily', size=12);
plt.title('Gender Equality in Carework by National Majority Religion', size=14);
plt.legend(religions, bbox_to_anchor=(1.05, .75))

# Save graph
plt.savefig('cara_religion.png')
```

### 4b. Gender Equality in Carework by National Majority Religion

```python
# Differences between work done by men and women
cd_orthodox_avg = cw_orthodox_avg - cm_orthodox_avg
cd_protestant_avg = cw_protestant_avg - cm_protestant_avg
cd_catholic_avg = cw_catholic_avg - cm_catholic_avg

# Convert to lists
cd_orthodox = cd_orthodox_avg.values.tolist()
cd_protestant = cd_protestant_avg.values.tolist()
cd_catholic = cd_catholic_avg.values.tolist()

# Plot lines on a graph
fig, ax = plt.subplots()
ax.plot(years, cd_orthodox, color='chocolate')
ax.plot(years, cd_protestant, color='mediumturquoise')
ax.plot(years, cd_catholic, color='darkmagenta')


# Add labels to the graph
plt.xlabel('Year',  size=12);
plt.ylabel('% More Women than Men Doing Carework Daily', size=9);
plt.title('Gender Equality in Carework by National Majority Religion', size=14);
plt.legend(religions, bbox_to_anchor=(1.05, .75))

# Save graph
plt.savefig('carb_religion.png')
```
