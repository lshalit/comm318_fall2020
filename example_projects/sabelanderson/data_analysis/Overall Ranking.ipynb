# Overall Ranking

## Set Up

```python
%matplotlib inline

import warnings
warnings.simplefilter('ignore')

import pandas as pd
import matplotlib.pyplot as plt
```

## Load Data

```python
# Load in big data set
df = pd.read_csv('../data/whole_GEI_dataset.csv')
```

```python
df
```

## Analyze

#### Countries ranked by GEI
________________________

```python
# Create dataframe for each year
y2005 = df[df['Year'] == 2005]
y2010 = df[df['Year'] == 2010]
y2012 = df[df['Year'] == 2012]
y2015 = df[df['Year'] == 2015]

# Sort from highest to lowest GEI score
y2005 = y2005.sort_values('Gender Equality Index', ascending = False)
y2010 = y2010.sort_values('Gender Equality Index', ascending = False)
y2012 = y2012.sort_values('Gender Equality Index', ascending = False)
y2015 = y2015.sort_values('Gender Equality Index', ascending = False)
```

```python
# Create ranked tables for each year with country and GEI
list2005 = y2005[['Country Long', 'Gender Equality Index']]
list2010 = y2010[['Country Long', 'Gender Equality Index']]
list2012 = y2012[['Country Long', 'Gender Equality Index']]
list2015 = y2015[['Country Long', 'Gender Equality Index']]
```

```python
# Create plots for each year
years = ['2005','2010','2012','2015']

# 2005
fig, ax = plt.subplots(figsize=(10,8))
ax.barh(y2005['Country Long'], y2005['Gender Equality Index'], color='lightblue')
plt.xlabel('Gender Equality Index',  size=12);
plt.ylabel('Country', size=12);
plt.title('2005 Gender Equality Index by Country', size=14);
plt.gca().invert_yaxis()

# Save graph
plt.savefig('gei_2005_rank.png')
```

```python
# 2010
fig, ax = plt.subplots(figsize=(10,8))
ax.barh(y2010['Country Long'], y2010['Gender Equality Index'], color='gainsboro')
plt.xlabel('Gender Equality Index',  size=12);
plt.ylabel('Country', size=12);
plt.title('2010 Gender Equality Index by Country', size=14);
plt.gca().invert_yaxis()

# Save graph
plt.savefig('gei_2010_rank.png')
```

```python
# 2012
fig, ax = plt.subplots(figsize=(10,8))
ax.barh(y2012['Country Long'], y2012['Gender Equality Index'], color='lightsteelblue')
plt.xlabel('Gender Equality Index',  size=12);
plt.ylabel('Country', size=12);
plt.title('2012 Gender Equality Index by Country', size=14);
plt.gca().invert_yaxis()

# Save graph
plt.savefig('gei_2012_rank.png')
```

```python
# 2015
fig, ax = plt.subplots(figsize=(10,8))
ax.barh(y2015['Country Long'], y2015['Gender Equality Index'], color='skyblue')
plt.xlabel('Gender Equality Index',  size=12);
plt.ylabel('Country', size=12);
plt.title('2015 Gender Equality Index by Country', size=14);
plt.gca().invert_yaxis()

# Save graph
plt.savefig('gei_2015_rank.png')
```
