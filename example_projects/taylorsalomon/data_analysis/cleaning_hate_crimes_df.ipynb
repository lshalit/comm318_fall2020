```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import re
```

# Import hate crimes data for New York State from data.gov

```python
hate_df=pd.read_csv('../data/Raw/Hate_Crimes_by_County_and_Bias_Type__Beginning_2010 (1).csv')
```

# Look at hate crimes data frame

```python
hate_df
```

* The data here includes hate crimes by county in NYS between 2010 and 2016, the groups that were targeted and whether the crime was against a person or property 

## I am only interested in looking at the hate crimes in each county in 2016 (election year)
* Make a new data frame called hate_2016_df that shows just the hate crimes that occurred in 2016

```python
hate_2016_df = hate_df[hate_df['Year']==2016]
```

## Check the shape of the data frame and what it looks like now

```python
hate_2016_df.shape
```

```python
hate_2016_df
```

## I want to look at total hate crimes per county (against people and property combined)
* Sum the "Total Incidents" column by county

```python
hate_2016_totals_df = hate_2016_df.groupby('County')['Total Incidents'].sum()
```

```python
hate_2016_totals_df 
```

* This is not a dataframe it is just a list of the number of incidents per county

## Make the summed totals of incidents per county a data frame

```python
hate_2016_totals_df = pd.DataFrame(hate_2016_totals_df)
```

```python
hate_2016_totals_df 
```

* The dataframe needs an index to be used later

## Index the dataframe

```python
hate_2016_totals_df2 = hate_2016_totals_df.reset_index()
```

```python
hate_2016_totals_df2
```

## Save the new data frame as a csv file in the data folder

```python
hate_2016_totals_df2.to_csv('../data/Hate_Crimes_clean.csv', index=False)
```

## Import cvs file of NYS population in order to add it to the data frame for comparison

```python
nys_pop_df = pd.read_csv('../data/Clean/NYSpop')
```

## Isolate data for the year 2016 to match our hate crimes data

```python
nys_pop_2016_df = nys_pop_df[nys_pop_df['Year'] == 2016]
```

```python
nys_pop_2016_df
```

* The geography column needs to be changed to match the one from the hate crimes data frame in order to be able to merge them later

## Rename the "Geography" column "County" so that it matches the column from the hate crimes data frame
* take the word "County" out of the name of each county so that it matches the county names in the hate crimes data frame

```python
nys_pop_2016_df['County'] = nys_pop_2016_df['Geography'].str.replace(' County', '')
```

## Isolate the columns we need to compare the population data frame to the hate crimes data frame later
* those would be population, county and FIPS code (because it may be needed later)

```python
nys_pop_2016_df2 = nys_pop_2016_df[['FIPS Code', 'Population', 'County']]
```

```python
nys_pop_2016_df2
```

## Save the clean NYS Population data frame as a csv file to be used in other notebooks

```python
nys_pop_2016_df2.to_csv('../data/NYS_Pop.csv', index = False)
```

## Merge the hate crimes data frame with the population data frame to create a new data frame to be used for comparisons

```python
hate_with_pop_df = hate_2016_totals_df2.merge(nys_pop_2016_df2)
```

## Create new column of normalized hate crime data based on county population
* Create an "HC_per_Mill" column of total hate incidents divided by the population and then multiplied by 1 million 

```python
hate_with_pop_df['HC_per_Mill'] = hate_with_pop_df['Total Incidents']/hate_with_pop_df['Population']*1000000
```

## Sort the values by HC_per_Mill to get an idea of where each county falls

```python
hate_with_pop_df.sort_values('HC_per_Mill')
```

## Save the final Hate Crimes data frame as a csv file to be used in other notebooks

```python
hate_with_pop_df.to_csv('../data/Hate_Crimes_Normalized.csv', index=False)
```
